```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

n = 1 #mol
a = 16.02 #(bar*L^2)/(mol^2)
R = 0.083144 #(bar*L)/(mol*K)
b = 0.1124 #L/mol

P,V = np.meshgrid(np.linspace(1,11,100), np.linspace(10,31,100))
T = ((P+a*(n**2/V**2))*(V-(n/b)))/(n*R)
plt.figure()
plt.imshow(T, cmap = 'inferno', vmin=50, vmax=2500)
plt.xlabel("Pressure")
plt.ylabel("Volume")
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N4 = np.random.uniform(0,1,10000)
inside = []
outside = []
for i in N4:
    x = np.random.uniform(0,1)
    y = np.random.uniform(0,1)
    if np.sqrt(x**2 + y**2) < 1:
        inside.append((x,y))
    else:
        outside.append((x,y))
        
pi_estimate = 4 * len(inside) / 10000
#print(pi_estimate)

plt.figure(figsize=(6,6))
plt.scatter([x[0] for x in inside], [x[1] for x in inside])
plt.scatter([x[0] for x in outside], [x[1] for x in outside])
plt.title("Estimate of Pi: "+str(pi_estimate))
plt.show()
```
